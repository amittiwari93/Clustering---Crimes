```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('crime_data.csv')
```

```python
dataset.shape
```

```python
dataset
```

### Data Cleaning

```python
dataset=dataset.drop(['Unnamed: 0'],axis=1)
dataset
```

```python
dataset.info()
```

### Scaling the features to common scale

```python
from sklearn.preprocessing import normalize
dataset_norm=pd.DataFrame(normalize(dataset),columns=dataset.columns)
```

```python
dataset_norm
```

### Using Dendogram to find optimal clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(dataset_norm, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

### Training the Hierarchy

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
```

```python
y_hc=pd.DataFrame(hc.fit_predict(dataset_norm),columns=['Cluster_Id'])
y_hc['Cluster_Id'].value_counts()
```

### Assign Cluster id to equivalent row data

```python
dataset_norm['Cluster_Id']=hc.labels_
```

```python
dataset_norm
```

```python
dataset_norm.groupby('Cluster_Id').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(dataset_norm['Cluster_Id'],dataset_norm['Murder'], c=hc.labels_) 
```
