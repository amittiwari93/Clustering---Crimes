```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('crime_data.csv')
```

```python
dataset.shape
```

```python
dataset
```

### Data Cleaning

```python
dataset=dataset.drop(['Unnamed: 0'],axis=1)
dataset
```

```python
dataset.info()
```

### Scaling the features to common scale

```python
from sklearn.preprocessing import StandardScaler
dataset_norm=StandardScaler().fit_transform(dataset)
dataset_norm
```

```python
dataset_norm
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=1,min_samples=4)
db.fit(dataset_norm)
```

```python
db.labels_
```

```python
dataset['Cluster_Id']=db.labels_
dataset
```

```python
dataset.groupby('Cluster_Id').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(dataset['Cluster_Id'],dataset['Murder'], c=db.labels_)
```
